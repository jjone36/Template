```python
import pandas as pd
```

```python
df = pd.read_csv('../input/pubg-finish-placement-prediction/train_V2.csv')
df = df.head(100)
```

```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
df.head()
```

## 1. Scatter plot 

```python
trace_1 = go.Scatter(x = df.maxPlace, y = df.matchDuration, mode = 'markers')
trace_2 = go.Scatter(x = df.maxPlace, y = df.walkDistance, mode = 'markers')
data = [trace_1, trace_2]
iplot(data)
```

## 2. Bar plot

```python
trace_1 = go.Bar(x = df.numGroups, y = df.teamKills, name = 'team kill')
trace_2 = go.Bar(x = df.numGroups, y = df.maxPlace/50, name = 'max place')

data = [trace_1, trace_2]
layout = go.Layout(barmode = 'group')
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

## 3. 3D Plot

```python
import plotly.figure_factory as ff
```

```python
x = df.numGroups
y = df.maxPlace
z = df.killPlace

trace = go.Scatter3d(x = x, y = y, z = z,
                    mode = 'markers', 
                    marker = dict(size = 12, color = z, opacity = .5))

data = [trace]
layout = go.Layout(margin = dict(l = 0, r = 0, b = 0, t = 0))
fig = go.Figure(data = data, layout = layout)

iplot(fig)
```

## 5. Maps

```python
import shapely
import shapefile
from plotly.figure_factory._county_choropleth import create_choropleth
import geopandas
```

```python
mn = pd.read_csv('../input/ex-map/minoritymajority.csv')
mn.head()
```

```python
scope = ['New York']
mn_sub = mn[mn.STNAME.isin(scope)]
mn_sub.head()
```

```python
values = mn_sub.TOT_POP.tolist()
fips = mn_sub.FIPS.tolist()
```

```python
colorscale = ["#171c42","#223f78","#1267b2","#4590c4","#8cb5c9","#b6bed5","#dab2be",
              "#d79d8b","#c46852","#a63329","#701b20","#3c0911"]
```

```python
fig = create_choropleth(scope = scope, values = values, fips = fips,
                        round_legend_values = True, 
                        # show_state_data = True,
                        simplify_county = 0,
                        simplify_state = 0,
                        county_outline = {'color': 'rgb(15, 15, 55)', 'width': .5},
                        state_outline = {'width' : .5},
                        legend_title = 'Population Per Country')
```

```python
iplot(fig, filename = 'Choropleth Map')
```
